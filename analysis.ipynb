```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('./datasets/combined_data_set3.csv')
random_dataset2_df = df.values

date_time = pd.to_datetime(df["Date_Time"])
hour = date_time.dt.hour
minute = date_time.dt.minute
#print(hour)
#print(minute)

df.head()
```

```python
bath_temp = df.drop(df[df['Bathroom_Temperature'] < 70].index)
bed1_temp = df.drop(df[df['Bedroom1_Temperature'] < 70].index)
bed2_temp = df.drop(df[df['Bedroom2_Temperature'] < 70].index)
kitch_temp = df.drop(df[df['Kitchen_Temperature'] < 70].index)
dine_temp = df.drop(df[df['Diningroom_Temperature'] < 70].index)
liv1_temp = df.drop(df[df['Livingroom1_Temperature'] < 70].index)
liv2_temp = df.drop(df[df['Livingroom2_Temperature'] < 70].index)
```

```python
plt.figure()
bath_temp.Bathroom_Temperature.plot()
bed1_temp.Bedroom1_Temperature.plot()
bed2_temp.Bedroom2_Temperature.plot()
kitch_temp.Kitchen_Temperature.plot()
dine_temp.Diningroom_Temperature.plot()
liv1_temp.Livingroom1_Temperature.plot()
liv2_temp.Livingroom2_Temperature.plot(figsize=(15,15))
plt.legend(loc='best')

ax = plt.gca()
ylim = ax.get_ylim()
plt.vlines(60, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(120, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(180, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(240, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(300, ylim[0], ylim[1], linestyles='dashed')

plt.title('Temperature Recorded During Random Scenario')
plt.xlabel('Minutes Elapsed')
plt.ylabel('Temperature\n(in Degrees\nFahrenheit)',labelpad=20).set_rotation(0)
plt.show()
```

```python
bath_hum = df.drop(df[df['Bathroom_Humidity'] > 100].index)
bed1_hum = df.drop(df[df['Bedroom1_Humidity'] > 100].index)
bed1_hum = bed1_hum.drop(df[df['Bedroom1_Humidity'] < 10].index)
bed2_hum = df.drop(df[df['Bedroom2_Humidity'] > 100].index)
kitch_hum = df.drop(df[df['Kitchen_Humidity'] > 100].index)
dine_hum = df.drop(df[df['Diningroom_Humidity'] > 100].index)
liv1_hum = df.drop(df[df['Livingroom1_Humidity'] > 100].index)
liv2_hum = df.drop(df[df['Livingroom2_Humidity'] > 100].index)
```

```python
plt.figure()
bath_hum.Bathroom_Humidity.plot()
bed1_hum.Bedroom1_Humidity.plot()
bed2_hum.Bedroom2_Humidity.plot()
kitch_hum.Kitchen_Humidity.plot()
dine_hum.Diningroom_Humidity.plot()
liv1_hum.Livingroom1_Humidity.plot()
liv2_hum.Livingroom2_Humidity.plot(figsize=(15,15))
plt.legend(loc='best')

ax = plt.gca()
ylim = ax.get_ylim()
plt.vlines(60, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(120, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(180, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(240, ylim[0], ylim[1], linestyles='dashed')
plt.vlines(300, ylim[0], ylim[1], linestyles='dashed')

plt.title('Humdity Recorded During Random Scenario')
plt.xlabel('Minutes Elapsed')
plt.ylabel('Humidity\n(in Percent)',labelpad=20).set_rotation(0)

plt.show()
```
